```python
import matplotlib.pyplot as plt
```

**Analyzing the ratio of Movies vs. TV Shows on Netflix (Pie Chart)**

```python
counts=df['type'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['lightgreen', '#FFFF99'])
plt.title('Distribution of Movie vs. TV Shows on Netflix')
plt.show()
```

**Insights**

The "70/30" Split: Netflix is predominantly a movie platform. 69.7% of the catalog consists of Movies, while 30.3% are TV Shows.

Implication: While Netflix is famous for "binge-worthy" series (like Stranger Things or Squid Game), the sheer volume of their library relies on single-viewing movies to keep users engaged between season releases.

```python
import seaborn as sns
```

**Counting Content by Rating (Histogram)**

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='rating', hue='rating', order=df['rating'].value_counts().index, palette='Reds_r', legend=False)

plt.title('Count pf content by Rating(TV-MA, PG-13, etc.)')
plt.xlabel('Rating Category')
plt.ylabel('Number of Titles')
plt.show()
```

**Insights**

Adult-Centric Catalog: The most common rating is TV-MA (Mature Audiences), followed by TV-14.

Not Just for Kids: Ratings like TV-Y (Young Children) and TV-G (General Audience) are very low in comparison.

Implication: Unlike Disney+, which focuses on family content, Netflixs core library is curated primarily for teens and adults.

**Tracking boom of content production over time (Stacked Histogram)**

```python
#Using seaborn
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='release_year', hue='type', multiple='stack', bins=50, palette='RdGy')

plt.title('Content Production Over the Years')
plt.xlabel('Year Released')
plt.ylabel('Count of Titles')
plt.xlim(2000, 2021)
plt.show()
```

```python
#Using Matplotlib
movie_years = df[df['type']=='Movie']['release_year']
tv_years = df[df['type'] == 'TV Show']['release_year']

plt.figure(figsize=(12, 6))

plt.hist([movie_years, tv_years], bins=50, stacked=True, color=['lightgreen', '#FFFF99'], label=['Movie', 'TV Show'], edgecolor='black', linewidth=0.5)

plt.title('Content Production Over the Years (Movie vs TV)')
plt.xlabel('Year Released')
plt.ylabel('Count of Titles')
plt.xlim(2000,2021)
plt.legend()
plt.show()
```

**Insights**

The Content Explosion (20152019): There was a massive exponential spike in content additions starting around 2015. This correlates with the "Streaming Wars" era where Netflix aggressively started producing Original content.

The COVID Effect (2020-2021): You can see a noticeable dip or plateau in 2020 and 2021. This is likely due to global production shutdowns during the pandemic, causing a lag in new releases.

**Ranking Top 10 Countries with most titles**

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values, hue=top_countries.index, legend=False, palette='Blues_r')

plt.title('Top 10 Countries with Most Titles on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

**Insights**

The United States dominates Netflixs library with the highest content contribution.

India is the second highest contributor, reflecting a strong regional content production market.

A large portion of titles are marked as Unknown country, which may indicate missing metadata.

**Ranking top 10 produced Genres**

```python
genres_list = df['listed_in'].str.split(', ').explode()

top_genres = genres_list.value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_genres.index, y=top_genres.values, hue=top_genres.index, legend = False, palette='Greens_r')

plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

**Insights**

International Movies and Dramas are the two most popular genres on the platform.

Comedies and International TV Shows also have strong presence  Netflix is shifting to global content strategy.

Genres like Documentaries and Action & Adventure have a decent presence, but niche categories rank lower.

**Key Takeaways**

Netflixs library is movie-dominant, but TV shows are rapidly growing.

US and India lead content production, reinforcing key market focus.

Netflix is heavily investing in international and drama content categories.

Mature-rated titles dominate, highlighting Netflixs target audience.

Production growth accelerated post-2015, with a dip around 2020 (COVID impact).
